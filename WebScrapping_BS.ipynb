```python
from bs4 import BeautifulSoup
import requests
```

```python
page1 = requests.get('https://www.pcmag.com/categories/processors/brands/intel')
page1
```

```python
soup1 = BeautifulSoup(page1.content)
```

```python
review1 = []

for i in soup1.find_all('p',class_="line-clamp-2 text-gray-600"):
    review1.append(i.text)

review1
```

```python
page2 = requests.get('https://www.pcmag.com/categories/processors/brands/intel?page=2')
page2
```

```python
soup2 = BeautifulSoup(page2.content)
```

```python
review2 = []

for i in soup2.find_all('p',class_="line-clamp-2 text-gray-600"):
    review2.append(i.text)

review2
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(review1)

df2 = pd.DataFrame(review2)
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1,df2])
```

```python
df.to_csv('intel_review.csv')
```
