```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
```

```python
page1 = requests.get('https://www.pcmag.com/categories/processors/brands/intel')
page1

soup1 = BeautifulSoup(page1.content)

product1 = []

for i in soup1.find_all('div',class_="font-stretch-ultra-condensed text-base font-semibold leading-compact md:text-xl"):
    product1.append(i.text)

d1 = pd.DataFrame(product1)
```

```python
page2 = requests.get('https://www.pcmag.com/categories/processors/brands/intel?page=2&_gl=1*1cqwhqr*_up*MQ..*_ga*MTQ3MDM0MzY3OS4xNzE4MTIzMjk2*_ga_2Y85WP1X8R*MTcxODEyMzI5Ni4xLjAuMTcxODEyMzI5Ni4wLjAuMA..')
page2

soup2 = BeautifulSoup(page2.content)

product2 = []

for i in soup2.find_all('div',class_="font-stretch-ultra-condensed text-base font-semibold leading-compact md:text-xl"):
    product2.append(i.text)

d2 = pd.DataFrame(product2)
```

```python
page3 = requests.get('https://www.pcmag.com/categories/processors/brands/intel?page=3')
page3

soup3 = BeautifulSoup(page3.content)

product3 = []

for i in soup3.find_all('div',class_="font-stretch-ultra-condensed text-base font-semibold leading-compact md:text-xl"):
    product3.append(i.text)

d3 = pd.DataFrame(product3)
```

```python
df = pd.concat([d1,d2,d3])

df
```

```python
len(df)
```

```python
d12 = pd.read_csv('/content/intel_review.csv')
d12
```

```python
da1 = d12.join(df)

da1
```

```python
da1.isnull().sum()
```

```python
da1.columns = ['Sr_No','Review','Product']
```

```python
da1
```

```python
from sklearn.impute import KNNImputer

from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()

da1['Product'] = lab_enc.fit_transform(da1['Product'])
```

```python
knnimp = KNNImputer(n_neighbors=2)
knn_imp = pd.DataFrame(knnimp.fit_transform(da1[['Product']]))

knn_imp
```

```python
da1['Product'] = knn_imp
```
